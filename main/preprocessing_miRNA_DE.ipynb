```python
import matplotlib.pyplot as plt
import numpy as np

# Define angles in degrees
angles_deg = [22.5, 30, 37.5, 45]  # Add more if needed
length = 1  # Length of each line

# Convert degrees to radians
angles_rad = [np.deg2rad(a) for a in angles_deg]

# Plot setup
fig, ax = plt.subplots(figsize=(6, 6))

# Draw each line from origin
for angle, deg in zip(angles_rad, angles_deg):
    x = length * np.cos(angle)
    y = length * np.sin(angle)
    ax.plot([0, x], [0, y], label=f'{deg}°', linewidth=2)

# Formatting
ax.set_xlim(0, 1.1)
ax.set_ylim(0, 1.1)
ax.set_aspect('equal')
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.legend(title="Angle")
ax.set_title("Lines at Various Angles from Origin")
plt.grid(True)
plt.tight_layout()
plt.show()
```
