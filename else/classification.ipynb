# Experiment on classifications

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example: gene expression data
# Rows: genes, Columns: time points
data = pd.DataFrame({
    'gene': ['gene1', 'gene2', 'gene3', 'gene4'],
    't1': [2.0, 3.0, 1.0, 5.0],
    't2': [3.5, 2.0, 1.0, 5.5],
    't3': [2.5, 2.5, 3.0, 6.0]
})

def classify_pattern(row):
    delta1 = row['t2'] - row['t1']
    delta2 = row['t3'] - row['t2']
    
    if delta1 > 0 and delta2 < 0:
        return 'Increase-then-Decrease'
    elif delta1 < 0 and delta2 > 0:
        return 'Decrease-then-Increase'
    elif delta1 > 0 and delta2 > 0:
        return 'Increasing'
    elif delta1 < 0 and delta2 < 0:
        return 'Decreasing'
    elif abs(delta1) < 0.2 and delta2 > 0:
        return 'Steady-then-Increase'
    elif abs(delta1) < 0.2 and delta2 < 0:
        return 'Steady-then-Decrease'
    else:
        return 'Unclassified'

data['pattern'] = data.apply(classify_pattern, axis=1)

# Plotting
for pattern in data['pattern'].unique():
    subset = data[data['pattern'] == pattern]
    for i, row in subset.iterrows():
        plt.plot(['t1', 't2', 't3'], row[['t1', 't2', 't3']], label=row['gene'])
    plt.title(f'Pattern: {pattern}')
    plt.xlabel('Time')
    plt.ylabel('Expression')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example gene expression data (replace this with your own DataFrame)
data = pd.DataFrame({
    'gene': ['gene1', 'gene2', 'gene3', 'gene4', 'gene5', 'gene6'],
    't1': [2.0, 3.0, 1.0, 5.0, 2.0, 2.0],
    't2': [3.5, 2.0, 1.0, 5.5, 2.1, 2.1],
    't3': [2.5, 2.5, 3.0, 6.0, 2.8, 1.5]
})

# Classification rule function
def classify_pattern(row, tol=0.0):
    delta1 = row['t2'] - row['t1']
    delta2 = row['t3'] - row['t2']
    
    if delta1 > tol and delta2 < -tol:
        return 'Increase-then-Decrease'
    elif delta1 < -tol and delta2 > tol:
        return 'Decrease-then-Increase'
    elif delta1 > tol and delta2 > tol:
        return 'Increasing'
    elif delta1 < -tol and delta2 < -tol:
        return 'Decreasing'
    elif abs(delta1) <= tol and delta2 > tol:
        return 'Steady-then-Increase'
    elif abs(delta1) <= tol and delta2 < -tol:
        return 'Steady-then-Decrease'
    else:
        return 'Unclassified'

# Apply classification
data['pattern'] = data.apply(classify_pattern, axis=1)

# Plot all genes with the same pattern in one plot
timepoints = ['t1', 't2', 't3']
unique_patterns = data['pattern'].unique()

for pattern in unique_patterns:
    subset = data[data['pattern'] == pattern]
    plt.figure(figsize=(8, 5))
    for _, row in subset.iterrows():
        plt.plot(timepoints, row[timepoints], marker='o', label=row['gene'])
    plt.title(f'Pattern: {pattern}')
    plt.xlabel('Time Point')
    plt.ylabel('Expression Level')
    plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```
