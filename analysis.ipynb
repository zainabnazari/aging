# mRNA and miRNA data for aging 2M 6M and 12M DEGs and main

```python
import pandas as pd
from pathlib import Path
import scipy.stats as stats
import time
```

```python
file_path = Path("/home/ebri/Documenti/final_anaysis_aging/data/")
file_path_r = Path("/home/ebri/Documenti/final_anaysis_aging/")
file_path_n = Path("/home/ebri/Documenti/final_anaysis_aging/data/new_version_miRNA")
```

```python
#mRNA

all_data_mRNA = pd.read_csv(file_path/"mRNA_mydata_filtered_Log2CPM_TMM_Norm_HtSeq_Union_paired_reads_counts_GTF_gene.txt",delimiter='\t')
all_data_mRNA.set_index('ID', inplace=True)


#changing the column names
all_data_mRNA.columns = all_data_mRNA.columns.str.extract(r'(^.*?_.*?)_S', expand=False)
all_data_mRNA.columns = all_data_mRNA.columns.astype(str).str.replace(r'^X', 'X_', regex=True)
```

```python
# Remove columns that contain "_2M_"
df_filtered_all_data_mRNA = all_data_mRNA.loc[:, ~all_data_mRNA.columns.str.contains(r'_2M_', regex=True)]
df_filtered_all_data_mRNA 
```

```python
# miRNA

all_data_miRNA = pd.read_csv(file_path_n/"only_6M_12M_miRNA_Bowtie1_align_miRBase_v_0_m_1_only_R1_samtools_idxstat_filtered_Log2CPM_TMM_Norm.txt",delimiter='\t')

all_data_miRNA.set_index('SampleID', inplace=True)
all_data_miRNA
```

```python
#change the column names
#all_data_miRNA.columns = all_data_miRNA.columns.str.replace(r'^NOGNA_', 'X_', regex=True)
all_data_miRNA.columns = all_data_miRNA.columns.str.replace(r'^(.*?)(_S\d+)$', r'X_\1', regex=True)

all_data_miRNA 
```

```python
# the correlation

# Ensure both dataframes have the same sample IDs (columns)
common_samples = df_filtered_all_data_mRNA .columns.intersection(all_data_miRNA.columns)
mRNA_filtered = df_filtered_all_data_mRNA [common_samples]
miRNA_filtered = all_data_miRNA[common_samples]

# Create a dataframe to store Pearson correlation results
correlation_results = []

# Compute Pearson correlation between each mRNA and each miRNA
for mRNA_gene in mRNA_filtered.index:
    for miRNA_gene in miRNA_filtered.index:
        # Get expression values across common samples
        mRNA_values = mRNA_filtered.loc[mRNA_gene]
        miRNA_values = miRNA_filtered.loc[miRNA_gene]

        # Compute Pearson correlation and p-value
        r, p = stats.pearsonr(mRNA_values, miRNA_values)

        # Store results
        correlation_results.append([mRNA_gene, miRNA_gene, r, p])

# Convert results to a DataFrame
correlation_df = pd.DataFrame(correlation_results, columns=["mRNA_Gene", "miRNA_Gene", "Pearson_r", "P_value"])

# Save results to a CSV file
correlation_df.to_csv("mRNA_miRNA_Pearson_Correlation.csv", index=False)

```

# DEGs Analysis Repetition 

```python
#mRNA

degs_data_mRNA = pd.read_csv(file_path/"only_6M_12M_mRNA_DEGs_FDR_5pc_1_5fc_filtered_Log2CPM_TMM_Norm_HtSeq_Union_paired_reads_counts_GTF_gene.txt",delimiter='\t')
degs_data_mRNA

degs_data_mRNA.set_index('SampleID', inplace=True)
degs_data_mRNA.columns = degs_data_mRNA.columns.str.replace(r'^(.*?)(_S\d+)$', r'X_\1', regex=True)
degs_data_mRNA
```

```python
#miRNA

degs_data_miRNA = pd.read_csv(file_path/"only_6M_12M_miRNA_DEGs_FDR_5pc_1_5fc_Bowtie1_align_miRBase_v_0_m_1_only_R1_samtools_idxstat_filtered_Log2CPM_TMM_Norm.txt",delimiter='\t')
degs_data_miRNA.set_index('SampleID', inplace=True)
degs_data_miRNA.columns = degs_data_miRNA.columns.str.replace(r'^(.*?)(_S\d+)$', r'X_\1', regex=True)
degs_data_miRNA
```

```python
start_time = time.time()

# the correlation

# Ensure both dataframes have the same sample IDs (columns)
common_samples = degs_data_mRNA .columns.intersection(degs_data_miRNA.columns)
mRNA_filtered = degs_data_mRNA[common_samples]
miRNA_filtered = degs_data_miRNA[common_samples]

# Create a dataframe to store Pearson correlation results
correlation_results = []

# Compute Pearson correlation between each mRNA and each miRNA
for mRNA_gene in mRNA_filtered.index:
    for miRNA_gene in miRNA_filtered.index:
        # Get expression values across common samples
        mRNA_values = mRNA_filtered.loc[mRNA_gene]
        miRNA_values = miRNA_filtered.loc[miRNA_gene]

        # Compute Pearson correlation and p-value
        r, p = stats.pearsonr(mRNA_values, miRNA_values)

        # Store results
        correlation_results.append([mRNA_gene, miRNA_gene, r, p])

# Convert results to a DataFrame
correlation_df = pd.DataFrame(correlation_results, columns=["mRNA_Gene", "miRNA_Gene", "Pearson_r", "P_value"])

# Save results to a CSV file
correlation_df.to_csv("DEGs_mRNA_miRNA_Pearson_Correlation.csv", index=False)
end_time = time.time()  # End time
elapsed_time = end_time - start_time  # Compute elapsed time
print(f"Execution Time: {elapsed_time:.4f} seconds")

```
